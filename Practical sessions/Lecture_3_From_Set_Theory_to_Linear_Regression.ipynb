ë‚´ì (Inner Product): $\langle x, y \rangle = x_1y_1 + x_2y_2 + ... + x_ny_n = ||x||||y||\cos(\theta)$

```python
# 1.1 ë²¡í„° ìƒì„± ë° ê¸°ë³¸ ì—°ì‚°
def vector_operations():
    """ë²¡í„° ê¸°ë³¸ ì—°ì‚° ì‹¤ìŠµ"""
    # ë²¡í„° ìƒì„±
    v1 = np.array([3, 4])
    v2 = np.array([1, 2])
    
    # ë²¡í„° ì—°ì‚°
    addition = v1 + v2
    scalar_mult = 2 * v1
    
    # ì§ì ‘ ë‚´ì  ê³„ì‚°
    dot_manual = sum(v1[i] * v2[i] for i in range(len(v1)))
    
    # NumPy ë‚´ì 
    dot_numpy = np.dot(v1, v2)
    
    print("ë²¡í„° v1:", v1)
    print("ë²¡í„° v2:", v2)
    print("\n--- ì—°ì‚° ê²°ê³¼ ---")
    print(f"ë§ì…ˆ: {addition}")
    print(f"ìŠ¤ì¹¼ë¼ ê³±: {scalar_mult}")
    print(f"ë‚´ì  (ì§ì ‘ ê³„ì‚°): {dot_manual}")
    print(f"ë‚´ì  (NumPy): {dot_numpy}")
    print(f"ë‘ ë°©ë²• ì¼ì¹˜: {dot_manual == dot_numpy}")
    
    return v1, v2

v1, v2 = vector_operations()
```

## ì‹¤ìŠµ 2: í–‰ë ¬ ì—°ì‚°ê³¼ ë³€í™˜

### ğŸ¯ í•™ìŠµ ëª©í‘œ
- í–‰ë ¬ ê³±ì…ˆì˜ ì´í•´ì™€ êµ¬í˜„

```python
# 2.1 í–‰ë ¬ ê³±ì…ˆ ì§ì ‘ êµ¬í˜„
def matrix_multiplication_comparison():
    """í–‰ë ¬ ê³±ì…ˆì„ ì§ì ‘ êµ¬í˜„í•˜ê³  NumPyì™€ ë¹„êµ"""
    
    # í–‰ë ¬ ìƒì„±
    A = np.array([[1, 2], [3, 4]])
    B = np.array([[5, 6], [7, 8]])
    
    # ì§ì ‘ êµ¬í˜„
    def manual_matmul(A, B):
        rows_A, cols_A = A.shape
        rows_B, cols_B = B.shape
        
        if cols_A != rows_B:
            raise ValueError("í–‰ë ¬ í¬ê¸°ê°€ ë§ì§€ ì•ŠìŠµë‹ˆë‹¤")
        
        C = np.zeros((rows_A, cols_B))
        
        for i in range(rows_A):
            for j in range(cols_B):
                for k in range(cols_A):
                    C[i][j] += A[i][k] * B[k][j]
        
        return C
    
    # ê³„ì‚°
    C_manual = manual_matmul(A, B)
    C_numpy = A @ B  # NumPy í–‰ë ¬ ê³±ì…ˆ
    
    print("í–‰ë ¬ A:")
    print(A)
    print("\ní–‰ë ¬ B:")
    print(B)
    print("\nì§ì ‘ êµ¬í˜„í•œ AÃ—B:")
    print(C_manual)
    print("\nNumPy AÃ—B:")
    print(C_numpy)
    print(f"\nê²°ê³¼ ì¼ì¹˜: {np.allclose(C_manual, C_numpy)}")
    
    return A

matrix_A = matrix_multiplication_comparison()
```

## ì‹¤ìŠµ 7: ì •ê·œë°©ì •ì‹ ì†”ë£¨ì…˜ (í•µì‹¬!)

### ğŸ¯ í•™ìŠµ ëª©í‘œ
- ì •ê·œë°©ì •ì‹ (X^T X)^(-1) X^T y ì§ì ‘ ê³„ì‚°
- NumPyë¡œ íš¨ìœ¨ì  êµ¬í˜„
- Scikit-learnê³¼ ê²°ê³¼ ë¹„êµ

```python
# 7.1 ì •ê·œë°©ì •ì‹ ìœ ë„ ë° êµ¬í˜„
def normal_equation_implementation():
    """ì •ê·œë°©ì •ì‹ì„ ë‹¨ê³„ë³„ë¡œ êµ¬í˜„"""
    
    # ë°ì´í„° ìƒì„±
    np.random.seed(42)
    n_samples = 100
    X = np.random.randn(n_samples, 1) * 3
    y = 2 + 3 * X + np.random.randn(n_samples, 1) * 0.5
    
    # Design matrix ìƒì„± (ì ˆí¸ í•­ ì¶”ê°€)
    X_with_intercept = np.c_[np.ones((n_samples, 1)), X]
    
    print("ì •ê·œë°©ì •ì‹: Î² = (X^T X)^(-1) X^T y")
    print("=" * 50)
    
    # ë‹¨ê³„ë³„ ê³„ì‚°
    print("\n1ë‹¨ê³„: X^T ê³„ì‚°")
    X_transpose = X_with_intercept.T
    print(f"   X shape: {X_with_intercept.shape}")
    print(f"   X^T shape: {X_transpose.shape}")
    
    print("\n2ë‹¨ê³„: X^T X ê³„ì‚°")
    XtX = X_transpose @ X_with_intercept
    print(f"   X^T X shape: {XtX.shape}")
    print(f"   X^T X:\n{XtX}")
    
    print("\n3ë‹¨ê³„: (X^T X)^(-1) ê³„ì‚°")
    XtX_inv = np.linalg.inv(XtX)
    print(f"   (X^T X)^(-1):\n{XtX_inv}")
    
    print("\n4ë‹¨ê³„: X^T y ê³„ì‚°")
    Xty = X_transpose @ y
    print(f"   X^T y shape: {Xty.shape}")
    
    print("\n5ë‹¨ê³„: Î² = (X^T X)^(-1) X^T y")
    beta = XtX_inv @ Xty
    print(f"   Î²:\n{beta}")
    
    print("\n" + "=" * 50)
    print(f"ìµœì¢… ê²°ê³¼: Î²â‚€ = {beta[0, 0]:.4f}, Î²â‚ = {beta[1, 0]:.4f}")
    
    return X, y, beta

X_data, y_data, beta_normal = normal_equation_implementation()
```

```python
# 7.2 ë‹¤ì–‘í•œ ë°©ë²•ìœ¼ë¡œ ê³„ì‚° ë¹„êµ
def compare_regression_methods():
    """ì •ê·œë°©ì •ì‹, NumPy, Scikit-learn ë¹„êµ"""
    
    print("ì„ í˜•íšŒê·€ ê³„ì‚° ë°©ë²• ë¹„êµ")
    print("=" * 60)
    
    # ë°©ë²• 1: ì •ê·œë°©ì •ì‹ ì§ì ‘ êµ¬í˜„
    X_with_intercept = np.c_[np.ones((len(X_data), 1)), X_data]
    beta_manual = np.linalg.inv(X_with_intercept.T @ X_with_intercept) @ X_with_intercept.T @ y_data
    print(f"\në°©ë²• 1 - ì •ê·œë°©ì •ì‹ ì§ì ‘ êµ¬í˜„:")
    print(f"  Î²â‚€ = {beta_manual[0, 0]:.6f}")
    print(f"  Î²â‚ = {beta_manual[1, 0]:.6f}")
    
    # ë°©ë²• 2: NumPyì˜ ìµœì†Œì œê³±ë²• (lstsq)
    beta_numpy, residuals, rank, s = np.linalg.lstsq(X_with_intercept, y_data, rcond=None)
    print(f"\në°©ë²• 2 - NumPy lstsq:")
    print(f"  Î²â‚€ = {beta_numpy[0]:.6f}")
    print(f"  Î²â‚ = {beta_numpy[1]:.6f}")
    
    # ë°©ë²• 3: Scikit-learn
    from sklearn.linear_model import LinearRegression
    reg = LinearRegression()
    reg.fit(X_data, y_data)
    print(f"\në°©ë²• 3 - Scikit-learn:")
    print(f"  Î²â‚€ = {reg.intercept_[0]:.6f}")
    print(f"  Î²â‚ = {reg.coef_[0, 0]:.6f}")
    
    # ê²°ê³¼ ë¹„êµ
    print("\n" + "=" * 60)
    print("ê²°ê³¼ ë¹„êµ:")
    print(f"  ë°©ë²• 1ê³¼ 2ì˜ ì°¨ì´: Î²â‚€={abs(beta_manual[0, 0] - beta_numpy[0]):.10f}, "
          f"Î²â‚={abs(beta_manual[1, 0] - beta_numpy[1]):.10f}")
    print(f"  ë°©ë²• 1ê³¼ 3ì˜ ì°¨ì´: Î²â‚€={abs(beta_manual[0, 0] - reg.intercept_[0]):.10f}, "
          f"Î²â‚={abs(beta_manual[1, 0] - reg.coef_[0, 0]):.10f}")
    
    print("\nâœ… ëª¨ë“  ë°©ë²•ì´ ë™ì¼í•œ ê²°ê³¼ë¥¼ ì‚°ì¶œí•©ë‹ˆë‹¤!")
    
    return beta_manual, beta_numpy, reg

beta1, beta2, sklearn_model = compare_regression_methods()
```
